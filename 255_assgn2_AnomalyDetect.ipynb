```python
dataset = pd.read_csv('Mall_Customers.csv') 
dataset.head()
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
data_train= dataset.sample(frac=0.8, random_state=786)
data_test = dataset.drop(data_train.index)

data_train.reset_index(drop=True, inplace=True)
data_test.reset_index(drop=True, inplace=True)

print('Train Data for Modeling: ' + str(data_train.shape))
print('Test Data For Predictions: ' + str(data_test.shape))
```

```python
from pycaret.anomaly import *
setup_anomaly_data = setup(data=dataset, normalize = True)
```

```python
iforest = create_model('iforest')
```

```python
plot_model(iforest)
```

```python
#plot_model(iforest, plot = 'umap')
```

```python
test_predict_IForest = predict_model(iforest, data = data_test)
print(test_predict_IForest)
```

```python
train_predict_IForest = predict_model(iforest, data = data_train)
train_predict_IForest.head()
```

```python
iforest_anomaly_rows = train_predict_IForest[train_predict_IForest['Anomaly'] == 1]
print(iforest_anomaly_rows.head())
```

```python
print(iforest_anomaly_rows.shape) 
```

```python
print(iforest_anomaly_rows.head())
```

```python
save_model(iforest,'Final_IForest_Model')
```

```python
saved_iforest = load_model('Final_IForest_Model')
```

```python
new_prediction = predict_model(saved_iforest, data=data_test)
```

```python
new_prediction.head()
```

```python
cluster = create_model('cluster', fraction = 0.025)
```

```python
plot_model(cluster)
```

```python
#plot_model(cluster, plot = 'umap')
```

```python
test_predict_Cluster= predict_model(cluster, data = data_test)
print(test_predict_Cluster)
```

```python
train_predict_Cluster = predict_model(cluster, data = data_train)
train_predict_Cluster.head()
```

```python
cluster_anomaly_rows = train_predict_Cluster[train_predict_Cluster['Anomaly'] == 1]
```

```python
print(cluster_anomaly_rows.head())
cluster_anomaly_rows.shape
```

```python
save_model(cluster,'Final_Cluster_Model')
```

```python
saved_cluster = load_model('Final_Cluster_Model')
```

```python
new__cluster_prediction = predict_model(saved_cluster, data=data_test)
print(new__cluster_prediction)
```
